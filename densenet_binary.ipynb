```python
import os
import sys
import csv
import numpy as np

multilabel = {}
with open("dataset/Data_Entry_2017.csv") as f:
    r = csv.reader(f)
    for i in r:
        if "|" in i[1]:
            multilabel[i[0]] = i[1]

multi_existed = {}
data_dir1 = "database/images/"
data_dir2 = "database/original_images/"
for i in os.listdir(data_dir1):
    if i in multilabel.keys():
        multi_existed[os.path.join(data_dir1,i)] = multilabel[i].split("|")    
for i in os.listdir(data_dir2):
    if i in multilabel.keys():
        multi_existed[os.path.join(data_dir2,i)] = multilabel[i].split("|")
multi_onehot = []

labels = ['Atelectasis', 'Cardiomegaly', 'Effusion', 'Infiltration', 'Mass', 'Nodule', 'Pneumonia', 
'Pneumothorax', 'Consolidation', 'Edema', 'Emphysema', 'Fibrosis', 'Pleural_Thickening', 'Hernia']
for k,v in multi_existed.items():
    indexs = []
    for i in v:
        indexs.append(labels.index(i))
    onehot = np.zeros(14)
    for i in indexs:
        onehot[i] = 1
    multi_onehot.append([k,onehot])
len(multi_onehot)    

with open ("dataset/multilabels.csv","w+") as f:
    w = csv.writer(f)
    for i in multi_onehot:
        w.writerow(i)
```

```python
from __future__ import print_function, division
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import time
import os
import seaborn as sns
import torch.nn.functional as F
use_gpu = torch.cuda.is_available()
torch.cuda.set_device(0)
from torchsummary import summary
from glob import glob
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
import pandas as pd
```

```python
use_gpu = torch.cuda.is_available()
device = torch.cuda.set_device(0)

trans = transforms.Compose([
            transforms.Resize(256),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])

class cheXnetDataset(Dataset):
    def __init__(self, csv_file, transform=None):
    
        self.frame = pd.read_csv(csv_file)
        self.transform = transform

    def __len__(self):
        return len(self.frame)

    def __getitem__(self, idx):
        img_name = self.frame.iloc[idx, 0]
        image = Image.open(img_name).convert("RGB")
        label = self.frame.iloc[idx, 1:].as_matrix()
        label = np.asarray(self.frame.iloc[idx,1:], dtype=np.float32)

    
        if self.transform:
            image = self.transform(image)

        return (image,label)
```

```python
batch_size = 8
train_dataset = cheXnetDataset(csv_file='dataset/twolabels_train.csv',transform=trans)
train_loader = torch.utils.data.DataLoader(train_dataset,shuffle=True,batch_size=batch_size, num_workers=4, pin_memory=True)
test_dataset = cheXnetDataset(csv_file='dataset/twolabels_test.csv',transform=trans)
test_loader = torch.utils.data.DataLoader(test_dataset,shuffle=True,batch_size=batch_size, num_workers=4, pin_memory=True)
dataloders = {'train':train_loader,'test':test_loader}
dataset_sizes = {'train':len(train_dataset),'test':len(test_dataset)}
dataset_sizes
```

```python
import warnings
warnings.filterwarnings('ignore')
densenet121 = models.densenet121(pretrained=False)
kernelCount = densenet121.classifier.in_features
densenet121.classifier = nn.Sequential(nn.Linear(kernelCount, 14), nn.Softmax())
if use_gpu:                 
    densenet121 = densenet121.cuda()       
densenet121 = densenet121.float()

checkpoint = torch.load("models/m-25012018-123527.pth.tar")
from collections import OrderedDict
new_state_dict = OrderedDict()
checkpoint["state_dict"]
for k, v in checkpoint["state_dict"].items():
    name = k[19:] # remove module.`
    name = name.replace("norm.1","norm1")
    name = name.replace("norm.2","norm2")
    name = name.replace("conv.1","conv1")
    name = name.replace("conv.2","conv2")
    new_state_dict[name] = v
densenet121.load_state_dict(new_state_dict)
for param in densenet121.parameters():
    param.requires_grad = False

densenet121.classifier.pop()
cls = nn.Sequential(nn.Linear(14, 2), nn.Softmax())


class chexnet(nn.Module):
    def __init__(self):
        super(chexnet, self).__init__()
        self.feature = densenet121
        self.cls = cls

    def forward(self, x):
        x = self.feature(x)
        #intervene here
#         print(x.shape)
#         x[:,2:4] = 0
#         x[:,12:14] = 0
        x = self.cls(x)
        return x

chexnet = chexnet().cuda()
```

```python
from tqdm import tqdm
from time import time
def train_model(model,criterion,optimizer,weight_name,num_epochs=10):
    since = time()
    best_model_wts = model.state_dict()
    best_acc = 0.0
    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)
        for phase in ['train', 'test']:
            if phase == 'train':
                model.train(True)  # Set model to training mode
            else:
                model.train(False)  # Set model to evaluate mode
            
            running_loss = 0.0
            running_corrects = 0.0
            
            for data in tqdm(dataloders[phase]):
                inputs, labels = data
                
                if use_gpu:
                    inputs = Variable(inputs.cuda())
                    labels = Variable(labels.cuda()).view(-1).long()
                else:
                    inputs, labels = Variable(inputs), Variable(labels)
                optimizer.zero_grad()
                outputs = model(inputs)

                _, preds = torch.max(outputs.data, 1)
                loss = criterion(outputs, labels)

                if phase == 'train':
                    loss.backward()
                    optimizer.step()   
                running_loss += loss.item()
                running_corrects += torch.sum(preds == labels.data)
            
            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = float(running_corrects) / float(dataset_sizes[phase])

            print('{} Loss: {:.10f} Acc: {:.10f}'.format(
                phase, epoch_loss, epoch_acc))

            # deep copy the model
            if phase == 'test' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = model.state_dict()
                state = {'model':model.state_dict(),'optim':optimizer.state_dict()}
                torch.save(state,weight_name)

    time_elapsed = time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best test Acc: {:4f}'.format(best_acc))

    # load best model weights
    model.load_state_dict(best_model_wts)
    return model
```

```python
optimizer = optim.Adam(filter(lambda p: p.requires_grad,chexnet.parameters()), lr=0.001)
criterion = nn.CrossEntropyLoss()
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
train_model(chexnet, criterion, optimizer,'chexnet_twoclass_intervene.pth',num_epochs=1)
```
